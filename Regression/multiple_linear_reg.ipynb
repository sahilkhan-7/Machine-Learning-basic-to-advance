```python
# Multiple linear regression
# Multiple linear regression is a type of regression analysis where the number of independent variables are more than one
# Example:
# Predicting the salary of an employee based on his position and level
# or 
# Predicting the price of a house based on its area, number of bedrooms, number of floors, etc.
```

### <font color = 'yellow'><b>Importing the libraries</b></font>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('D:\Python work\Machine learning revision\Regression\housing\house_sales.csv', sep = '\t')
```

```python
subset = ['AdjSalePrice', 'SqFtTotLiving', 'SqFtLot', 'Bathrooms', 'Bedrooms', 'BldgGrade']
df[subset].head()
```

```python
predictors = ['SqFtTotLiving', 'SqFtLot', 'Bathrooms', 'Bedrooms', 'BldgGrade']
outcome = 'AdjSalePrice'
house_lr = LinearRegression()
house_lr.fit(df[predictors], df[outcome])
```

```python
print(f'Intercept: {house_lr.intercept_}')
print(f'Coefficients: {house_lr.coef_}')
for name, coef in zip(predictors, house_lr.coef_):
    print(f'{name}: {coef}')
```

```python
y_pred = house_lr.predict(df[predictors])
y_pred
```

```python
rmse = np.sqrt(mean_squared_error(df[outcome], y_pred))
r2 = r2_score(df[outcome], y_pred)
print(f'RMSE: {rmse}')
print(f'R^2: {r2}')
```

```python
import statsmodels.api as sm
```

```python
# x_with_constant = sm.add_constant(df[predictors])
model = sm.OLS(df[outcome], df[predictors].assign(const=1))
```

```python
results = model.fit()
print(results.summary())
```

## <font color = 'Red'><b>Model Selection and Stepwise Regression</b></font>

```python
predictors = ['SqFtTotLiving', 'SqFtLot', 'Bathrooms', 'Bedrooms',
              'BldgGrade', 'PropertyType', 'NbrLivingUnits',
              'SqFtFinBasement', 'YrBuilt', 'YrRenovated', 
              'NewConstruction']

X = pd.get_dummies(df[predictors], drop_first=True, dtype=int)
X['NewConstruction'] = [1 if nc else 0 for nc in X['NewConstruction']]

house_full = sm.OLS(df[outcome], X.assign(const=1))
results = house_full.fit()
print(results.summary())
```
