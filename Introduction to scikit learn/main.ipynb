### Supervised learning example : Simple Linear Regression

```python
rng = np.random.RandomState(42) # This line creates a random number generator with a fixed seed value of 42
x = 10*rng.rand(50)
y = 2*x - 1 + rng.randn(50)
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)
model
```

```python
X = x[:, np.newaxis]
X.shape
```

```python
model.fit(X, y)
```

```python
model.coef_ # Slope of the linear regression line
```

```python
model.intercept_ # intercept of the data
```

```python
xfit = np.linspace(-1, 11)
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
```

```python
plt.scatter(X, y)
plt.plot(xfit, yfit)
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state=1)
```

```python
y_model = modelNB.predict(Xtest)
```
