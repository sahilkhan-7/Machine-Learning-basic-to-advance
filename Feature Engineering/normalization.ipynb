```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class label', axis=1),
                                                    df['Class label'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

# MinMaxScaler from scratch

```python
class min_max_scaler:
    def __init__(self) -> None:
        self.mins = list()
        self.maxs = list()
        self.cols = list()
        
    def fit(self, train_df):
        self.mins = train_df.min().values
        self.maxs = train_df.max().values
        self.cols = train_df.columns
        
    def transform(self, test_df):
        if len(test_df.columns) != len(self.cols):
            raise ValueError("Number of columns in the test data must match the number of columns in the training data.")
        
        test_df = test_df.copy()
        for i, col in enumerate(test_df.columns):
            test_df[col] = (test_df[col] - self.mins[i])/(self.maxs[i] - self.mins[i])
        return test_df
    
    def fit_transform(self, train_df):
        self.fit(train_df)
        return self.transform(train_df)
```

```python
# Scikit Learn's MinMaxScaler 
scikit_scaler = MinMaxScaler()

scikit_scaler.fit(df.drop('Class label', axis=1))

scikit_scaled_df = scikit_scaler.transform(df.drop('Class label', axis = 1))

scikit_scaled_df = pd.DataFrame(scikit_scaled_df, columns = df.drop('Class label', axis = 1).columns)
scikit_scaled_df
```
