```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
```

```python
df = sns.load_dataset('mpg')
```

```python
df.sample(10)
```

```python
df.iloc[(np.arange(0, df.shape[0], step=20))]
```

```python
df.origin.value_counts()
```

```python
# disproportionate stratified sampling
df.groupby('origin', group_keys = False).apply(lambda x: x.sample(3))
```

```python
# Proportionate stratified sampling
df.groupby('origin', group_keys = False).apply(lambda x: x.sample(frac=0.05))
```

```python
sns.displot(df['displacement'], kind='kde')
```

Central Limit Theorem

```python
sample_mean_list = list()
for i in range(0, 50):
    random_sample = df['displacement'].sample(50)
    sample_mean_list.append(random_sample.mean())
    
sample_mean_list
```

```python
sns.displot(sample_mean_list, kind='kde')
```

```python
df['displacement'].mean(), statistics.mean(sample_mean_list)
```

```python
# Doing same for mpg

sns.displot(df['mpg'], kind='kde')
```
