```python
from scipy.stats import poisson, expon
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
l = 4
s = 7
poisson.pmf(s, l)
```

```python
poisson.pmf(120, 100)
```

```python
# Generating poisson distribution with sample size 10000 

x = poisson.rvs(mu = 3, size = 10000)
x
```

```python
plt.hist(x, density=True, ec = 'k')
```

```python
x = expon.rvs(scale = 40, size = 10000)
x
```

```python
sns.distplot(x)
```
